# Generar datos

```python
n = 20
```

```python
x = np.ones((n,2))
x[:,1] = linspace(1.0,20.0,n)
y = x[:,1]*1.5 + np.random.normal(size = n, scale = 5.0)
```

```python
plot(x[:,1],y,'bo')
```

# Definir modelo

```python
def modelo_lineal(theta,x):
    return np.dot(theta,np.transpose(x))
```

# Inicializar par√°metros

```python
theta = np.array([0.01,0.01])
```

```python
alpha = 0.0000001
```

# Entrenar modelo

```python
for i in range(10000):
    theta[0] = theta[0] + alpha*np.sum((y - modelo_lineal(theta, x))*x[:,0])
    theta[1] = theta[1] + alpha*np.sum((y - modelo_lineal(theta, x))*x[:,1])
```

```python
theta
```

```python
plot(x[:,1],y,'bo')
plot(theta[0] + theta[1]*x[:,1],'r-')
```

```python
m = dot(transpose(x),x)
```

```python
l = dot(transpose(x),y)
```

```python
theta_an = dot(linalg.inv(m),l)
```

```python
plot(x[:,1],y,'bo')
plot(theta[0] + theta[1]*x[:,1],'r-')
plot(theta_an[0] + theta_an[1]*x[:,1],'r--')
```
