```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = tf.linspace(0.0,2.0*np.pi,100)
```

```python
@tf.function
def seno(x):
    y = tf.math.sin(x)
    y_x = tf.gradients(y, x)
    return tf.gradients(y_x, x)
```
